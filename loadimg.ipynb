```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Load the image
img = mpimg.imread('images/abra.png')
plt.imshow(img)
plt.show()
```

```python
import numpy as np

# Convert to grayscale manually
gray_img = np.dot(img[..., :3], [0.2989, 0.5870, 0.1140])

# Display the grayscale image
plt.imshow(gray_img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
gray_img[gray_img == 0] = 1
plt.imshow(gray_img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
from scipy import ndimage
import imageio
```

```python
# Load the image
image = imageio.imread('images/abra.png', mode='F')

```

```python



# Apply Sobel filter in the x and y directions
sobel_x = ndimage.sobel(image, axis=0)
sobel_y = ndimage.sobel(image, axis=1)

# Combine the Sobel filters
sobel_combined = np.hypot(sobel_x, sobel_y)

# Display the result
plt.imshow(sobel_combined, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Invert the colors
inverted_sobel = 1 - sobel_combined

# Display the inverted image
plt.imshow(inverted_sobel, cmap='gray')
plt.axis('off')
plt.show()
```
